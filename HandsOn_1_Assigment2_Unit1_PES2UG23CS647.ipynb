```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")
```

# Initialize Text Generation Pipeline

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
#PES2UG23CS647
prompt = (
    "Zarathos is a legendary character in a dark fantasy world. "
    "Born under a cursed eclipse, Zarathos grew up to become"
)
```

# Generating Backstory

```python
output = generator(
    prompt,
    max_length=200,
    num_return_sequences=1,
    temperature=0.9,
    top_p=0.95
)

print(output[0]["generated_text"])
```

```python
#PES2UG23CS647
#Cleaner Formatting
story = output[0]["generated_text"]

print("🧙 Character Bio: Zarathos\n")
print(story)
```
